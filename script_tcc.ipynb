```python
import string

import pandas as pd
import numpy as np
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from rake_nltk import Rake
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from unidecode import unidecode
```

```python
# Flag para ativar o pre-processamento
aply_pre_process = True
```

```python
def pre_process(corpus, stop_words_remv = True, lemmatizetion = True, stemmization = True):
    corpus = corpus.lower()
    stopset = None
    if (stop_words_remv):
        stopset = stopwords.words('english')
        stopset += list(string.punctuation)

    if (lemmatizetion):
        lemmatizer = WordNetLemmatizer()
        corpus = " ".join([lemmatizer.lemmatize(i) for i in word_tokenize(corpus) if i not in stopset])
    
    if (stemmization):
        stemmer = PorterStemmer()
        corpus = " ".join([stemmer.stem(i) for i in word_tokenize(corpus)])
    
    # remove non-ascii characters
    corpus = unidecode(corpus)
    
    return corpus
```

```python
# descartando as vírgulas entre os nomes completos dos atores e obtendo apenas os três primeiros nomes
df['Actors'] = df['Actors'].map(lambda x: x.split(',')[:3])

# colocando os gêneros em uma lista de palavras
df['Genre'] = df['Genre'].map(lambda x: x.lower().split(','))

df['Director'] = df['Director'].map(lambda x: x.split(' '))
```

Mesclar o nome e o sobrenome de cada ator e diretor, por isso é considerado como uma palavra e não há mistura entre pessoas que compartilham um primeiro nome

```python
for index, row in df.iterrows():
    row['Actors'] = [x.lower().replace(' ','') for x in row['Actors']]
    row['Director'] = ''.join(row['Director']).lower()
```

```python
if (aply_pre_process):
    for index, row in df.iterrows():
        plot = row['Plot']
        row['Plot'] = pre_process(plot)
```

```python
# iniciando uma nova coluna
df['Key_words'] = ""
```

```python
for index, row in df.iterrows():
    plot = row['Plot']
    
    # instanciando Rake, por padrão usa palavras irrelevantes em inglês do NLTK 
    # e descarta todos os caracteres de pontuação
    r = Rake()

    # extraindo as palavras passando o texto
    r.extract_keywords_from_text(plot)

    # obtendo o dicionário com palavras-chave e suas pontuações
    key_words_dict_scores = r.get_word_degrees()
    
    # atribuindo as palavras-chave à nova coluna
    row['Key_words'] = list(key_words_dict_scores.keys())
```

```python
# descartando a coluna Plot
df.drop(columns = ['Plot'], inplace = True)

df.set_index('Title', inplace = True)
df.head()
```

```python
df['bag_of_words'] = ''
columns = df.columns
for index, row in df.iterrows():
    words = ''
    for col in columns:
        if col != 'Director':
            words = words + ' '.join(row[col])+ ' '
        else:
            words = words + row[col]+ ' '
    row['bag_of_words'] = words
    
df.drop(columns = [col for col in df.columns if col!= 'bag_of_words'], inplace = True)

df.head()
```

```python
# instanciando e gerando a matriz de contagem
count = CountVectorizer()
count_matrix = count.fit_transform(df['bag_of_words'])
```

```python
# criando uma Série para os títulos dos filmes para que sejam associados a uma 
# lista numérica ordenada que usarei posteriormente para corresponder aos índices
indices = pd.Series(df.index)
indices[:5]
```

```python
# gerando a matriz de similaridade de cosseno
cosine_sim = cosine_similarity(count_matrix, count_matrix)
cosine_sim
```

```python
# função que recebe o título do filme como entrada e retorna os 10 principais filmes recomendados
def recommendations(title, cosine_sim = cosine_sim):
    
    recommended_movies = []
    
    # obtendo o índice do filme que corresponde ao título
    idx = indices[indices == title].index[0]

    # criando uma série com as pontuações de similaridade em ordem decrescente
    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending = False)

    print(score_series)
    # obtendo os índices dos 10 filmes mais parecidos
    top_10_indexes = list(score_series.iloc[1:11].index)
    
    # preenchendo a lista com os títulos dos 10 melhores filmes correspondentes
    for i in top_10_indexes:
        recommended_movies.append(list(df.index)[i])
        
    return recommended_movies
```

```python
def recommender_10_more_frequenly(movies, cosine_sim = cosine_sim):
    more_frequent = {}
    # Para cada filme recebido pegar as recomendações dele e colocar em um dicionário com uma contagem
    for movie in movies:
        recommended_movies = recommendations(movie, cosine_sim)
        for i in recommended_movies:
            if i not in more_frequent:
                more_frequent[i] = 1
            else:
                more_frequent[i] += 1
    
    # cria um dicionário ordenado de forma crescente com os filmes mais recomendados
    more_frequent = dict(sorted(more_frequent.items(), key=lambda item: item[1], reverse=True))

    # remover do dicionário os filmes passados na lista
    for movie in movies:
        if movie in more_frequent:
            del more_frequent[movie]

    #Pega os 10 mais frequentes
    ten_more_flequent_name = list(more_frequent.keys())[:10]

    return ten_more_flequent_name
```

```python

def get_similarity(interacted_items, cosine_similarity = cosine_sim):
        
    # obtendo os índices dos filmes interagidos
    idcs_interact = []
    for item in interacted_items:
        idcs_interact.append(indices[indices == item].index[0])

    similarity_non_interacted_items = []

    # obtendo os índices dos filmes não interagidos
    idcs_not_interact = [idc for idc in list(indices.index) if idc not in idcs_interact]

    # pegando lista de similaridades dos itens não interagidos
    for idc in idcs_not_interact:
        similarity_non_interacted_items.append(cosine_similarity[idc])

    # zerando a similaridade dos itens interagidos nas listas de similaridade dos itens não interagidos
    for similarity_item in similarity_non_interacted_items:
        for index in idcs_interact:
            similarity_item[index] = 0.0

    # obtendo a similaridade média
    num_of_items = len(similarity_non_interacted_items[0])
    sum_similarity = np.array([0.0] * num_of_items)
    for similarity_item in similarity_non_interacted_items:
        sum_similarity += np.array(similarity_item)

    average_similarity = sum_similarity / num_of_items

    items_more_similar = pd.Series(average_similarity).sort_values(ascending = False)

    # pegando o top 10 indices de filmes mais similares
    top_10_indexes = list(items_more_similar.iloc[0:10].index)

    recommended_movies = []
    
    # preenchendo a lista com os títulos dos 10 melhores filmes correspondentes
    for i in top_10_indexes:
        recommended_movies.append(list(df.index)[i])
        
    return recommended_movies
```

```python
movies = ['Interstellar',
 'Gladiator',
 '2001: A Space Odyssey',
 'Star Wars: The Force Awakens',
 'Catch Me If You Can',
 'Baby Driver',
 'The Godfather: Part II',
 'The Wolf of Wall Street',
 'On the Waterfront',
 'Goodfellas']

movies_recomended = get_similarity(movies)
movies_recomended
```
